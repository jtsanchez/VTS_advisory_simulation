```python
def preprocess_function(text):

    # Convert text to lowercase
    if text is not None and isinstance(text, str):
        text = text.lower()
    
    # retain text after second colon only
    split_text = text.split(':', 2)  # Split the text at the second colon
    if len(split_text) > 2:
        cleaned_text = split_text[2].strip()  # Get the text after the second colon and remove leading/trailing whitespace
    else:
        cleaned_text = text  # If there are less than two colons, keep the original text

    return cleaned_text

# apply preprocessing function
vis_line_outages['processed_message'] = vis_line_outages['MESSAGE'].apply(preprocess_function)
vis_line_outages
```
